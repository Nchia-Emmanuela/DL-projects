```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.dpi'] = 150
```

```python
# Define the function and its derivative
def fxn(x):
    return 3*x**2 - 3*x +4

def df_deriv(x):
    return 6*x - 3
```

```python
# Creating points between -2 and 2
x = np.linspace(-2, 2, 2001)

# Plotting the function
plt.plot(x, fxn(x), x, df_deriv(x), label='f(x) and f\'(x)')
plt.xlim(x[[0, -1]])
plt.grid()
plt.title('Gradient Descent in 1D')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
# Gradient Descent Implementation

def gradient_descent(start, learning_rate, iterations):
    # Random starting point
    localmin = start
    for i in range(iterations):
        gradient = df_deriv(localmin)
        localmin = localmin - learning_rate * gradient
    return localmin
```

```python
localmin = np.random.choice(x, 1)
learning_rate = 0.1
training_epochs = 100

localmin = gradient_descent(localmin, learning_rate, training_epochs)
print(f"Local minimum occurs at x = {localmin}, f(x) = {fxn(localmin)}")
```

```python
plt.plot(x, fxn(x), x, df_deriv(x))
plt.plot(localmin, df_deriv(localmin), 'ro')
plt.plot(localmin, fxn(localmin), 'ro')

plt.xlim(x[[0, -1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['f(x)', 'df', 'f(x) min'])
plt.title('Empirical local minimum: %s'%localmin[0])
plt.show
```
