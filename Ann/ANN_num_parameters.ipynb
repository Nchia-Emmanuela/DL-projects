```python
import numpy as np
import torch
import torch.nn as nn
```

```python
# Build two simple models
widenet = nn.Sequential(
    nn.Linear(2, 4),  # hidden layer
    nn.Linear(4, 3)     # Output layer
)
# N:B the nn.Linear layer is the transformation matrix layer that gets us from the input to the output. The first argument
#  represents the number of input features, while the second argument represets the number of output features to be fed into the next layer.
# N:B the nn.Linear layer is a linear transformation: y = xA^T + b, where A is the weight matrix and b is the bias vector. The first argument is the number of input features, and the second argument is the number of output features.
deepnet = nn.Sequential(
    nn.Linear(2,2), # hidden layer
    nn.Linear(2,2), # hidden layer
    nn.Linear(2,3)  # Output layer
)

print("Widenet: ", widenet)
print('')
print("Deepnet: ", deepnet)
```

```python
for param in deepnet.named_parameters():
    print(param)
    print('')
```

```python
# Count total number of nodes (By counting the number of biases)
total_nodes_in_widenet = 0
for param in widenet.named_parameters():
    print(param)
    if 'bias' in param[0]: # if the parameter is a bias
        total_nodes_in_widenet += len(param[1]) # add the number of nodes in that layer to the total count

total_nodes_in_deepnet = 0
for param_name, param_vect in deepnet.named_parameters():
    if 'bias' in param_name: # if the parameter is a bias
        total_nodes_in_deepnet += len(param_vect) # add the number of nodes in that layer to the total count

print(f"Total number of nodes in widenet: {total_nodes_in_widenet}")
print(f"Total number of nodes in deepnet: {total_nodes_in_deepnet}")
```

```python
# named_parameters returns a tuple of (name, parameter)
# parameter returns a tensor of the parameter, so it actually same as the named_parameters without the name
for param in widenet.parameters():
    print(param)
    print('')
```

```python
# Count total number of trainable parameters

number_of_nodes_in_widenet = 0
for param in widenet.parameters():
    if param.requires_grad:
        number_of_nodes_in_widenet += param.numel() # add the number of nodes in that layer to the total count

number_of_nodes_in_deepnet = 0
for param in deepnet.parameters():
    if param.requires_grad:
        number_of_nodes_in_deepnet += param.numel() # add the number of nodes in that layer to the total count

print(f"Total number of nodes in widenet: {number_of_nodes_in_widenet}")
print(f"Total number of nodes in deepnet: {number_of_nodes_in_deepnet}")
```

```python
# Using list comprehension
number_of_nodes_in_widenet = sum([param.numel() for param in widenet.parameters() if param.requires_grad])
number_of_nodes_in_deepnet = sum([param.numel() for param in deepnet.parameters() if param.requires_grad])

print(f"Total number of nodes in widenet: {number_of_nodes_in_widenet}")
print(f"Total number of nodes in deepnet: {number_of_nodes_in_deepnet}")
```
