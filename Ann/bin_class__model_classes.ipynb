```python
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F

import matplotlib.pyplot as plt
import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
```

```python
# Create Data

num_samples = 100
blur = 1
A = [1, 1]
B = [5,1]

# generate data
a = [A[0] + np.random.randn(num_samples) *blur, A[1] + np.random.randn(num_samples) *blur]
b = [B[0] + np.random.randn(num_samples) *blur, B[1] + np.random.randn(num_samples) *blur]

# True labels
labels_np = np.vstack((np.zeros((num_samples, 1)), np.ones((num_samples, 1))))

# concatenate into a matrix
data_np = np.hstack((a, b)).T

# Convert to PyTorch tensors
data = torch.tensor(data_np).float()
labels = torch.tensor(labels_np).float()

# Visualize the data
figure = plt.figure(figsize=(5, 5))
plt.plot(data[np.where(labels == 0)[0], 0], data[np.where(labels == 0)[0], 1], 'bs', label='Class A')
plt.plot(data[np.where(labels == 1)[0], 0], data[np.where(labels == 1)[0], 1], 'ko', label='Class B')
plt.title('Data Visualization')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
# Create Model using classes
class ANNClass(nn.Module):
    def __init__self(self):
        super(ANNClass, self).__init__()
        self.fc1 = nn.Linear(2, 1)
        self.fco = nn.Linear(1, 1)

    def forward(self, x):

        # Pass through input layer
        x = self.fc1(x)

        # apply relu
        x = F.relu(x)

        # output layer
        x = self.fco(x)
        x = torch.sigmoid(x)

        return x
# Initialize class 
ANNClassify = ANNClass()
```
